```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_normal(N=1000, mu=10_000, sigma=1000):
    return np.random.normal(mu, sigma, N).astype(int)
```

```python
create_normal()[:20]
```

```python
np.logspace(1, 6, 6).astype(int)
```

```python
E_MAXIMUM_N = 8
PRECISION = 2
xs = np.logspace(1, E_MAXIMUM_N, PRECISION*E_MAXIMUM_N).astype(int)
ys = []
stddevs = []
for x in xs:
    data = create_normal(N=x)
    tr = %timeit -n3 -r3 -o pass; np.quantile(data, q=.3)
    ys.append(tr.best)
    stddevs.append(tr.stdev)
```

```python
plt.errorbar(xs, ys, stddevs)
plt.xlabel("N ")
plt.ylabel("Time (seconds)")
plt.xscale("log")
plt.yscale("log")
plt.title("Logarithmic time complexity of np.quantile")
plt.savefig("time_compl.svg")
```
